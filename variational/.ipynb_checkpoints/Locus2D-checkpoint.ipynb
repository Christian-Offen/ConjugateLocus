```python
from EndPtJAX import Locus, ContFun
import jax.numpy as jnp
from jax import ops
import matplotlib.pyplot as plt
from tqdm import tqdm   # progress bar
import scipy.io         # Matlab export
```

```python
# sample values for 2d ellipsoid in R3
n = 3 # dimension outer space
b = jnp.array([0.9,1.2,1.6]) # ellipsoid coefficients
T = 1 # time
N=10 # steps
dt=T/N # discretisation parameter
XStart = jnp.array([0,0]) # start point of geodesic map
```

```python
l2 = Locus(n,b,T,N,XStart) # Create 2d geodesic problem (object)
```

```python
# domain patch
nx = 40
ny = 40
l = jnp.pi*1.1/N # pi/N is approximation to injectivity radius
x_ = jnp.linspace(XStart[0]-l,XStart[0]+l,num=nx)
y_ = jnp.linspace(XStart[1]-l,XStart[1]+l,num=ny)
xx,yy = jnp.meshgrid(x_,y_)
```

```python
# evaluate over domain patch
@jnp.vectorize
def LocusChartVec(x,y):
    return l2.LocusChart(jnp.array([x,y]))

Val = LocusChartVec(xx,yy)
```

```python
# Plot critical set of "endpt".
# Every second level sets is related to the charts and does not relate to conjugate points
figlevel=plt.figure()
c = plt.contour(xx, yy, Val, levels=[0])
figlevel.axes[0].set_xlabel('x0')
figlevel.axes[0].set_ylabel('x1')
```

```python
# Map all critical sets to ellipsoid
nc = jnp.shape(c.collections[0].get_paths())[0]

# initalise figure locus
fig4 = plt.figure(figsize=(10, 8))

# use pseudo-arclength continuation to improve the level sets
# steps for continuation
M = 40
ds = 0.05

for k in range(0,nc):
    
    verts=c.collections[0].get_paths()[k].vertices
    
    # improve vertices of level set
    cL = jnp.zeros((M,2))
    cL=ops.index_update(cL,ops.index[0],verts[1])
    cL=ops.index_update(cL,ops.index[1],verts[0])

    for j in tqdm(range(2,M)):
        
        cL=ops.index_update(cL,ops.index[j],ContFun(cL[j-2],cL[j-1],l2.LocusChart,ds))
    
    cL = cL[2:M+1,:]
    
    vertsX=jnp.array(list(map(l2.endptChart,cL)))
    vertsX = jnp.transpose(vertsX)   
        
    # Plot critical set and locus 
    ax4 = fig4.add_subplot(nc,2,2*k+1)
    ax5 = fig4.add_subplot(nc,2,2*k+2)
    ax4.plot(vertsX[0],vertsX[1])
    ax5.plot(cL[:,0],cL[:,1])
    ax4.set_xlabel('x0')
    ax4.set_ylabel('x1')
    ax5.set_xlabel('x0')
    ax5.set_ylabel('x1')
    
```

```python
# Replot favourite locus

M = 40
ds = 0.05

verts=c.collections[0].get_paths()[4].vertices

cL = jnp.zeros((M,2))
cL=ops.index_update(cL,ops.index[0],verts[1])
cL=ops.index_update(cL,ops.index[1],verts[0])

for j in tqdm(range(2,M)):
    cL=ops.index_update(cL,ops.index[j],ContFun(cL[j-2],cL[j-1],l2.LocusChart,ds))
    
cL = cL[2:M+1,:]
vertsX=jnp.array(list(map(l2.endptChart,cL)))
vertsX = jnp.transpose(vertsX)
plt.plot(vertsX[0],vertsX[1])
```

```python
# data ellipsoid
u = jnp.linspace(0, 2 * jnp.pi, 100)
v = jnp.linspace(0, jnp.pi, 200)
x = 1/jnp.sqrt(b[0]) * jnp.outer(jnp.cos(u), jnp.sin(v))
y = 1/jnp.sqrt(b[1]) * jnp.outer(jnp.sin(u), jnp.sin(v))
z = 1/jnp.sqrt(b[2]) * jnp.outer(jnp.ones(jnp.size(u)), jnp.cos(v))

vertsX = jnp.transpose(vertsX)
vertsX=jnp.array(list(map(l2.chartF,vertsX)))
vertsX = jnp.transpose(vertsX)


# plot conjugate locus on ellipsoid
figLc = plt.figure()
axLc = figLc.add_subplot(111,projection='3d')

a=[0, 100, 155, 200]

axLc.plot_surface(x[a[0]:a[1],a[2]:a[3]],y[a[0]:a[1],a[2]:a[3]],z[a[0]:a[1],a[2]:a[3]],alpha=0.5)
axLc.plot(vertsX[0],vertsX[1],vertsX[2])
axLc.set_xlabel('x0')
axLc.set_ylabel('x1')
axLc.set_zlabel('x2')
#axLc.view_init(3, 30)
```

```python
# Matlab export
scipy.io.savemat('./Data/Locus2D.mat', dict(lc=vertsX,x=x,y=y,z=z,ref=l2.chartS(XStart)))
```
